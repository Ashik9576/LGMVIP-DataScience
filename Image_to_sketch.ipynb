## **Image to Pencil Sketch with Python**

**Importing Libraries**

```python
import cv2 as cv
#import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image
import matplotlib.pylab as plt

```

```python
# mounting drive
from google.colab import drive
drive.mount('/content/drive')
```

**Loading Image**

```python
img = cv2.imread('/content/drive/MyDrive/ashik.jpg')
```

**Converting the origianl image to greyscale**

```python
#convertinng BGR to RGB
gray_img = cv2.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray_img)
```

**Inverting Grey scale image**

```python
invert = cv.bitwise_not(gray_img)
cv2_imshow(invert)
```

**Blurred Inverted image**

```python
blur= cv.GaussianBlur(invert,(21,21),0)
cv2_imshow(blur)
```

**Inverted Blur Image**

```python
inverted_blur= cv.bitwise_not(blur)
cv2_imshow(inverted_blur)
```

**Creating Image to Pencil Sketch**

```python
sketch_filter = cv.divide(gray_img, inverted_blur, scale=255.0)
cv2_imshow(sketch_filter)
```
